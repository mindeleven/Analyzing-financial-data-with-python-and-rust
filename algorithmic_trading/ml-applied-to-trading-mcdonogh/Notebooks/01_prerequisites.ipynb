### Performance Evaluation – Confusion Matrix

<img src="../assets/images/confusion_matrix_1.png" width="80%" />

*Screenshot taken from the Udemy Couse [Machine Learning Applied to Stock & Crypto Trading](https://www.udemy.com/course/machine-learning-applied-to-stock-crypto-trading-python/) by Shaun McDonogh.*

A very useful tool in both binary and multiclass classification is the ***confusion matrix***.

- True Positive (bottom right): we predict a one and get a one (one=market goes up)

- True Negative (top left): we predict a zero and get a zero (zero=market goes down)

__TRUE means we got them right.__

- False Positive (top right): a false positive is where we predicted a one but it was actually a zero

- False Negative (bottom left): a false negative is where we predicted a zero but it was actually a one

__FALSE means we got things wrong.__

The confusion matrix can help us to get insights into how is your model making predictions or where is it leaning towards. If the predictions of the model turn out to be false, we can ask
- Why might it be predicting that?
- Have we given it the right data?
- Have we given it the right size of data?
- Have we, for example, only given it data where the market's gone up and therefore it's learned just to predict up?

<img src="../assets/images/confusion_matrix_2.png" width="80%" />

*Screenshot taken from the Udemy Couse [Machine Learning Applied to Stock & Crypto Trading](https://www.udemy.com/course/machine-learning-applied-to-stock-crypto-trading-python/) by Shaun McDonogh.*

Based on a confusion matrix we can calculate some metrics:

1. __Accuracy:__ How many did I get correct in total versus the total amount of predictions or available data that there is? Accuracy is very straightforward but not very useful for us.

2. __Precision:__ Precision takes the true positives as a percentage of the true positives and the false positives. What it's basically saying is, what percentage out of all of my predictions of a one (TRUE) did I get right? Where I'm wrong her, __I will lose money__, therefore I I want that percentage to be as high as possible. ***When we get into supervised machine learning we will optimise xgboost, I will to aim for improving precision.***

3. __Recall:__ What are my true positives as a percentage of my true positives and false negatives? The recall is basically looking at how many of the pool did you get. Out of everything I could have predicted, how many of them did I catch? Missed opportunities means you missed positive trades but __you didn't lose money__.

4. __F1 score:__ The F1 score can be seen as the true accuracy. It looks at the precision multiplied by the recall over the precision plus the recall multiplied by two (2 x ((Precision x Recall) / (Precision + Recall))) and takes like a balanced view of the precision and recall.

### Working with Returns

```python
# add a column
df["Returns"] = df["Close"].pct_change() # calculatig oercentage change for the close column
df.head(3)
```

```python
# drop Na
df.dropna(inplace=True) # inplace=True overwrites the dataFrame
df.head(3)
```

```python
# calculate Log Returns
# log returns are important because whenever you're summing returns for a certain asset
# like when we're doing back testing or equity curves, we'll need log returns, the natural log return to be specific
# shift(1) means we're dividing the close price by the close price one row before
df["lrets"] = np.log(df["Close"] / df["Close"].shift(1))
df.head(3)
```

> __2. Why It's Useful in Finance__<br/>
> In financial calculations, the natural logarithm is preferable to simple percentage returns for several important reasons:
> 
> - Symmetry of Returns: When you use natural logarithmic returns (also called log returns), the mathematical properties create a more symmetric representation of gains and losses. With percentage returns, a 50% gain followed by a 50% loss does not bring you back to the original value, but log returns solve this asymmetry.
> 
> - Compounding Behavior: Log returns can be easily added together to calculate multi-period returns. This makes them superior for analyzing cumulative performance over time, especially when dealing with multiple investment periods.
> 
> - Statistical Properties: Log returns are more normally distributed, which makes them more suitable for statistical analysis and financial modeling. This characteristic is crucial for risk management and portfolio theory.

> __3. Calculation Example__<br/>

> - Percentage Return: (110 - 100) / 100 = 10%
> - Log Return: ln(110/100) = ln(1.1) ≈ 0.0953 or 9.53%
> 
> The log return provides a more mathematically consistent representation of the price change.

> __4. Practical Application__<br/>

> - Calculating portfolio performance
> - Developing statistical models

```python
# cumulative sum of the log returns
# CumSum Log Returns
df["lrets_cumsum"] = df["lrets"].cumsum()
df.head()
```

```python
# Normalise Log Returns
# to get the true returns, the true cumulative sum returns, we need to normalize it with the exponents
df["rets_normal"] = np.exp(df["lrets_cumsum"]) - 1
df.head()
```

```python
# filling the NaNs instead of droppimg them
df.fillna(0, inplace=True)
df.head(3)
```

### Conditionals

```python
# add conditional statements
# df["Close"].shift(-1) is one step in the future
# is close of 'one step in the future' greater than today's close?
# then creating a target column and setting it's vaule to one if condition is met
# the TARGET column is the future we want to predict
df_new.loc[df["Close"].shift(-1) > df["Close"], "TARGET"] = 1
df_new.loc[df["Close"].shift(-1) <= df["Close"], "TARGET"] = -1
df_new.head()

# this could also be achieved with numpy's np.where:
# create the TARGET column
# df["TARGET"] = np.where(df["Close"].shift(-1) > df["Close"], 1, -1)
```
