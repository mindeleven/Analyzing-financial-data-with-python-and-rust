```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
#plt.style.use("seaborn-v0_8")
```

```python
# specifying the the first two rows are headers of a multi-index with header=[0, 1]
# specifying the Date column as an indes coulmn with index_col = 0
# transforming the index to datetime index with parse_dates = [0]
df = pd.read_csv("../../assets/data/multi_assets.csv", header = [0, 1], index_col = 0, parse_dates = [0])
df
```

#### __Focusing on the Close Prices__

```python
# copying Close prices in another data frame with copy()
close = df.Close.copy() # select daily close prices only
close
```

```python
close.describe()
```

```python
# creating a price chart for a single instrument (Boeing); dropping missing values first
close.BA.dropna().plot(figsize = (15, 8), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

```python
# creating a price chart for all six instruments; doesn't look too good
close.dropna().plot(figsize = (15, 8), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

#### We're comparing absolute prices here which are on a different scale and therefore are hard to compare.
#### A higher Price does not imply a higher value or a better performance.
