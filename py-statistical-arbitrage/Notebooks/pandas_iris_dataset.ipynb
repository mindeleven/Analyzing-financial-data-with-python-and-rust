### Displaying specific rows using `iloc` and `loc` functions

```python
# loc() uses the index name of the row to display a particular row
data.loc[data["Species"] == "Iris-setosa"].head(10)
```

```python
# iloc() uses the index of a row to display the data in the row's columns
data.iloc[5] # shows data of the row with index 5 in the example above
```

```python
data["Species"].value_counts() # counts the number of times a particular instance or data has occurred
```

### Descriptive statistics

```python
# some descriptive statistic
sum = data["SepalLengthCm"].sum()
mean = data["SepalLengthCm"].mean()
median = data["SepalLengthCm"].median()
print("Sum:",sum, "\nMean:", mean, "\nMedian:",median) 
```

```python
min = data["SepalLengthCm"].min() 
max = data["SepalLengthCm"].max() 
print("Minimum:",min, "\nMaximum:", max) 
```

### Altering the dataset

```python
# adding a column to the dataset, in this example "total_values"
# this row should store the sum of all integer values
# (1) extracting the columns which have integer values
cols = data.columns
print(cols)
```

```python
# (2) taking that columns that have integer values
cols = cols[1:5]
# (3) saving this columbs it a new dataframe variable 
data_temp = data[cols] 
```

```python
data_temp
```

```python
# calculating the sum over the columns of data_temp
data_temp[cols].sum(axis=1) # to set the sum function to rows we us "axis=1" as an argument
# for columns we would use "axis=0"
```

```python
# adding new column "total_values" to the dataframe data with the calculation we used above
data["total_values"] = data_temp[cols].sum(axis=1)
```

```python
# renaming the column
# creating a dictionary first to update our new column names
newcols = {
    "Id": "id", 
    "SepalLengthCm": "sepallength",
    "SepalWidthCm": "sepalwidth"
}
data.rename(columns=newcols,inplace=True) 
print(data.head())
```

### Visualzing data with Seaborn

For the plotting part I'm following the examples in the Kaggle article [Seaborn plot to visualize Iris data](https://www.kaggle.com/code/rakesh6184/seaborn-plot-to-visualize-iris-data).

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = pd.read_csv(data_dir + "Iris.csv")
```

```python
iris.head()
```

```python
iris.drop('Id',axis=1,inplace=True)
```

```python
iris.info() # checking if there are any missing values
```

```python
iris['Species'].value_counts()
```

```python
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=iris,size=5) # joint plot example
```

```python
# FacetGrid plot with the help of matplotlib
%matplotlib inline
sns.FacetGrid(iris,hue='Species')\
.map(plt.scatter,'SepalLengthCm','SepalWidthCm')\
.add_legend()
```

```python
sns.boxplot(x='Species',y='PetalLengthCm',data=iris) # boxplot example
```

```python
ax=sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='gray') # strip plot
```

```python
# combining box and strip plots
ax=sns.boxplot(x='Species',y='SepalLengthCm',data=iris)
ax=sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='gray')
```

```python
sns.violinplot(x='Species',y='SepalLengthCm',data=iris) # violin plot
```

```python
sns.pairplot(data=iris,kind='scatter') # pair plot
```

```python
sns.pairplot(iris,hue='Species')
```
