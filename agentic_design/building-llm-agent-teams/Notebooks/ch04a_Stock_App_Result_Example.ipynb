- **Forward P/E**: Similar to the P/E ratio but based on future earnings estimates. This metric provides a forward-looking perspective on valuation.
- **Price to Book**: Compares a company's market value to its book value, providing insight into market perceptions of the company�s growth potential.
- **Debt to Equity (D/E)**: Measures a company's financial leverage. A high D/E ratio indicates significant debt, which can be risky if not managed well.
- **Return on Equity (ROE)**: Indicates how well a company utilizes its equity to generate profits. A higher ROE is generally favorable.
